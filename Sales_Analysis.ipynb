# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

sns.set(rc={'figure.figsize':(5,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Define variable
x = df['Age'].to_frame()  #independent variable
y = df['Orders']      #dependent variable
```

```python
from sklearn.model_selection import train_test_split
# split the dataset into 80% for training and 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
m = reg.coef_[0]
c = reg.intercept_
print(f"Slope / Regression co-efficient m = {m} \n\t y-intercept c = {c}")
```

```python
X = x.to_numpy()
y_pred = m*X + c
```

```python
plt.scatter(x,y, color='blue')
plt.plot(X, y_pred, c='g')
plt.xlabel('Orders')
plt.ylabel('Age')
plt.title('LinearRegression with slope')
plt.show()
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r_squared_score = r2_score(y_test, y_pred)
print("R-squared score = ", r_squared_score)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=50)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
